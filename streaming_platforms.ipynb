```python
import pandas as pd
```

```python
platforms_csv = "Resources/streaming_platforms.csv"
platforms_df = pd.read_csv(platforms_csv)
platforms_df.head()
```

```python
# drop unnecessary columns
platforms_df.drop('Unnamed: 0', 1, inplace=True)
platforms_df.drop('Age', 1, inplace=True)
platforms_df.drop('Type', 1, inplace=True)
platforms_df.dropna(inplace=True)
```

```python
# new dataframe to hold column separated values
directors_df = platforms_df["Directors"].str.split(",", n = -1, expand = True)

# add new columns to platforms_df
platforms_df['Director 1'] = directors_df[0]
platforms_df['Director 2'] = directors_df[1]

# drop old 'Directors' column
platforms_df.drop(columns = ['Directors'], inplace=True)
platforms_df.head()
```

```python
# new dataframe to hold comma separated columns
country_df = platforms_df['Country'].str.split(',', n = -1, expand = True)

# add new columns to platforms_df (only taking first 2)
platforms_df['Country 1'] = country_df[0]
platforms_df['Country 2'] = country_df[1]

# drop old 'Country' column
platforms_df.drop(columns = ['Country', 'ID'], inplace=True)
platforms_df.head()
```

```python
# new dataframe to hold comma separated columns
genres_df = platforms_df['Genres'].str.split(',', n = -1, expand = True)

# add new columns to platforms_df (only taking first 3)
platforms_df['Genre 1'] = genres_df[0]
platforms_df['Genre 2'] = genres_df[1]
platforms_df['Genre 3'] = genres_df[2]

# drop old 'Genres' column
platforms_df.drop(columns = ['Genres'], inplace=True)
platforms_df.head()
```

```python
# new dataframe to hold comma separated columns
language_df = platforms_df['Language'].str.split(',', n = -1, expand = True)

# add new columns to platforms_df (only taking first 2)
platforms_df['Language 1'] = language_df[0]
platforms_df['Language 2'] = language_df[1]

# drop old 'Language' column
platforms_df.drop(columns = ['Language'], inplace=True)
platforms_df.head()
```
